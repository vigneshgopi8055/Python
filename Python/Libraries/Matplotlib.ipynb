# MATPLOTLIB

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.

* Create publication quality plots.
* Make interactive figures that can zoom, pan, update.
* Customize visual style and layout.
* Export to many file formats.
* Embed in JupyterLab and Graphical User Interfaces.
* Use a rich array of third-party packages built on Matplotlib.
* matplotlib consists of barplot,line,scatter etc

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Lineplot

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.plot(x,y)
```

### Barplot

```python
x = [1,2,3,4,5]
y = [5,6,7,8,9]
plt.bar(x,y)
plt.xlabel = ('X-AXIS')
plt.ylabel = ('Y-AXIS')
plt.title = ('Matplotlib')
```

### Histogram

```python
x = [1,2,2,3,3,4,6,7,1,8,2,6]
plt.hist(x)
#shows the no of repeated elements
```

### Scatterplot 

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.scatter(x,y)
plt.plot(x,y)
plt.bar(x,y)
plt.xlabel = ('X-AXIS')
plt.ylabel = ('Y-AXIS')
plt.title = ('Matplotlib')
```

### SubPlot 

it takes 3 arguements
* row
* column
* index

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.plot(data['age'],data['bmi'])

plt.subplot(1,2,2)
plt.plot(data['age'],data['expenses'])
```

```python
plt.figure(figsize=(10,6))
plt.subplot(3,3,1)
plt.plot(data['age'],data['bmi'])

plt.subplot(3,3,2)
plt.plot(data['age'],data['expenses'])

plt.subplot(3,3,3)
plt.plot(data['age'],data['expenses'])

plt.subplot(3,3,4)
plt.plot(data['age'],data['expenses'])

plt.subplot(3,3,5)
plt.plot(data['age'],data['expenses'])
```

```python
#fig is the frame of plot
#ax is the axis
fig,ax = plt.subplots(2,2)
ax[0,0].plot(range(10),'r')
ax[0,1].plot(range(10),'b')
ax[1,0].plot(range(10),'g')
ax[1,1].plot(range(10),'y')
```

```python
year = [2010,2011,2012,2013,2014,2015]
yeild_apples = [20,50,10,60,20,100]
yeild_orange = [90,30,40,20,50,30]
plt.plot(year,yeild_apples,marker = 'o')
plt.plot(year,yeild_orange,marker = 'x')
plt.xlabel = ('year')
```

```python
year = [2010,2011,2012,2013,2014,2015]
yeild_apples = [20,50,10,60,20,100]
yeild_orange = [90,30,40,20,50,30]
plt.plot(year,yeild_apples,'s-b')
plt.plot(year,yeild_orange,'o--r')
plt.xlabel = ('year')
```
